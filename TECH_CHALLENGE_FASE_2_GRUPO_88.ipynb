#IMPORTAÇÃO DE BIBLIOTECAS

```python
import pandas as pd
import numpy as np
from math import radians, sin, cos, sqrt, atan2
```

## Construção do DataFrame

```python
# ============================================================
# Matriz de Distâncias entre Capitais Brasileiras
# Para uso no Google Colab
# ============================================================


# ------------------------------------------------------------------
# 1. Coordenadas geográficas das capitais (latitude, longitude)
# ------------------------------------------------------------------
capitais = {
    "AC - Rio Branco":      (-9.9754, -67.8249),
    "AL - Maceió":          (-9.6658, -35.7350),
    "AP - Macapá":          (0.0349, -51.0694),
    "AM - Manaus":          (-3.1190, -60.0217),
    "BA - Salvador":        (-12.9714, -38.5014),
    "CE - Fortaleza":       (-3.7172, -38.5433),
    "DF - Brasília":        (-15.7797, -47.9297),
    "ES - Vitória":         (-20.3155, -40.3128),
    "GO - Goiânia":         (-16.6869, -49.2648),
    "MA - São Luís":        (-2.5297, -44.3028),
    "MT - Cuiabá":          (-15.6010, -56.0974),
    "MS - Campo Grande":    (-20.4697, -54.6201),
    "MG - Belo Horizonte":  (-19.9167, -43.9345),
    "PA - Belém":           (-1.4558, -48.5044),
    "PB - João Pessoa":     (-7.1195, -34.8450),
    "PR - Curitiba":        (-25.4278, -49.2731),
    "PE - Recife":          (-8.0476, -34.8770),
    "PI - Teresina":        (-5.0892, -42.8019),
    "RJ - Rio de Janeiro":  (-22.9068, -43.1729),
    "RN - Natal":           (-5.7945, -35.2110),
    "RS - Porto Alegre":    (-30.0346, -51.2177),
    "RO - Porto Velho":     (-8.7612, -63.9004),
    "RR - Boa Vista":       (2.8235, -60.6758),
    "SC - Florianópolis":   (-27.5954, -48.5480),
    "SP - São Paulo":       (-23.5505, -46.6333),
    "SE - Aracaju":         (-10.9091, -37.0677),
    "TO - Palmas":          (-10.2491, -48.3243),
}

# ------------------------------------------------------------------
# 2. Função Haversine para calcular distância entre dois pontos
# ------------------------------------------------------------------
def haversine(lat1, lon1, lat2, lon2):
    """
    Calcula a distância em quilômetros entre dois pontos
    na superfície da Terra usando a fórmula de Haversine.
    """
    R = 6371.0  # Raio médio da Terra em km

    phi1, phi2 = radians(lat1), radians(lat2)
    dphi       = radians(lat2 - lat1)
    dlambda    = radians(lon2 - lon1)

    a = sin(dphi / 2)**2 + cos(phi1) * cos(phi2) * sin(dlambda / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return round(R * c, 1)

# ------------------------------------------------------------------
# 3. Construir a matriz de distâncias
# ------------------------------------------------------------------
estados = list(capitais.keys())
n = len(estados)

matriz = np.zeros((n, n))

for i, estado_i in enumerate(estados):
    lat1, lon1 = capitais[estado_i]
    for j, estado_j in enumerate(estados):
        lat2, lon2 = capitais[estado_j]
        matriz[i, j] = haversine(lat1, lon1, lat2, lon2)

df_distancias = pd.DataFrame(matriz, index=estados, columns=estados)
df_distancias = df_distancias.astype(int)  # distâncias em km inteiros

# ------------------------------------------------------------------
# 4. Exibir o dataframe
# ------------------------------------------------------------------
print(f"Matriz {n}x{n} — distâncias em km (linha: origem | coluna: destino)\n")

# Configurações de exibição para o Colab
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows",    None)
pd.set_option("display.width",       None)
pd.set_option("display.float_format", "{:,.0f}".format)

display(df_distancias)   # display() renderiza melhor no Colab; use print() se preferir

# ------------------------------------------------------------------
# 5. Análises rápidas
# ------------------------------------------------------------------
print("\n" + "="*60)
print("ANÁLISES RÁPIDAS")
print("="*60)

# Par mais próximo (excluindo diagonal zero)
df_upper = df_distancias.copy().astype(float)
np.fill_diagonal(df_upper.values, np.nan)

min_val  = df_upper.stack().min()
max_val  = df_upper.stack().max()
min_pair = df_upper.stack().idxmin()
max_pair = df_upper.stack().idxmax()

print(f"\n📍 Capitais mais próximas : {min_pair[0]}  ↔  {min_pair[1]}")
print(f"   Distância              : {min_val:,.0f} km")

print(f"\n📍 Capitais mais distantes: {max_pair[0]}  ↔  {max_pair[1]}")
print(f"   Distância              : {max_val:,.0f} km")

# Distância média entre todas as capitais
media = df_upper.stack().mean()
print(f"\n📊 Distância média entre capitais: {media:,.0f} km")

# Capital mais central (menor soma de distâncias)
soma_por_capital = df_distancias.sum(axis=1)
capital_central  = soma_por_capital.idxmin()
print(f"\n🎯 Capital mais central (menor soma de distâncias): {capital_central}")
print(f"   Soma total de distâncias: {soma_por_capital[capital_central]:,.0f} km")

# ------------------------------------------------------------------
# 6. (Opcional) Exportar para Excel
# ------------------------------------------------------------------
# df_distancias.to_excel("distancias_capitais_brasil.xlsx")
# print("\nArquivo 'distancias_capitais_brasil.xlsx' salvo com sucesso.")
```
